```python
def extract_temp(s):
    if 'AAXX' in s:
        
        temp_idx = 29
    else:
        temp_idx = 18
    if len(s) >=temp_idx + 5:        
        s = s[temp_idx:temp_idx+5]

        
        if s.isdigit() and s[0] in ['1']:
            if s[1] == '0':
                return float(s[2:]) / 10
            elif s[1] == '1':
                return -float(s[2:]) /10
        else:
            return np.nan
    else:
        return np.nan
            
            
```
